```python
import os
import time
import json
from collections import defaultdict
from PIL import Image

# Carica un'immagine
test_images = "PPE_Detection 2.v1i.yolov8-obb_FINALE/test/images"

image_path = [os.path.join(test_images, fname) for fname in os.listdir(test_images) if fname.endswith(('.png', '.jpg', '.jpeg'))]

save_dir = 'output/YOLOv8/test_set_predictions_results'
os.makedirs(save_dir, exist_ok=True)
names = ['Gloves', 'Hat', 'Non-Gloves', 'Non-Hat', 'Non-Shoes', 'Vest', 'Worker', 'Shoes', 'Non-Vest']

for img_path in image_path:
    img = Image.open(img_path)
    
    # Inizia il timer per l'elaborazione dell'immagine
    start_time = time.time()
    
    # Esegui il modello
    results = model(img)
    
    # Ferma il timer e calcola il tempo di elaborazione per l'immagine corrente
    elapsed_time = time.time() - start_time
    
    detection_summary = results[0].__str__()
    base_name = os.path.splitext(os.path.basename(img_path))[0]
    txt_file_path = os.path.join(save_dir, f'{base_name}.txt')

    for result in results:
        result.plot()

    # Salvo i risultati
    # Estrarre le informazioni necessarie
    for i, result in enumerate(results):
        # Dizionario per mantenere il conteggio totale per ogni classe
        class_counts = defaultdict(int)
        detections_with_confidence = []

        for box in result.boxes:
            cls = int(box.cls)  # Classe come intero
            class_name = model.names[cls]  # Nome della classe
            class_counts[class_name] += 1  # Incrementa il conteggio per questa classe
            
            # Aggiungi la confidenza per ogni rilevamento
            confidence = float(box.conf)
            detections_with_confidence.append({
                "class": class_name,
                "confidence": confidence
            })

        # Crea il dizionario di riepilogo delle detections, includendo il tempo di elaborazione e le confidenze
        detection_summary = {
            "image": f"{base_name}",
            "detections": dict(class_counts),
            "detections_with_confidence": detections_with_confidence,  # Aggiungi confidenza
            "processing_time_seconds": elapsed_time  # Aggiungi il tempo di elaborazione
        }

        # Salva il risultato in un file JSON
        base_name = os.path.splitext(os.path.basename(img_path))[0]
        json_file_path = os.path.join(save_dir, f'{base_name}.json')
        with open(json_file_path, 'w') as json_file:
            json.dump(detection_summary, json_file, indent=4)
```

```python
import os
import json

mapping = [
    ("worker", "worker"),
    ("vest", "vest"),
    ("shoes", "shoes"),
    ("gloves", "gloves"),
    ("helmet", "helmet"),
]

def filter_greater_than_zero(data):
    return {key: value for key, value in data.items()}

# Funzione per leggere i file di ground truth
def read_ground_truth(file_path):
    ground_truth = {}
    with open(file_path, 'r') as f:
        for line in f:
            elements = line.split()
            class_id = int(elements[0])
            if class_id == 0:
                ground_truth["gloves"] = ground_truth.get("gloves", 0) + 1
            elif class_id == 1:
                ground_truth["helmet"] = ground_truth.get("helmet", 0) + 1
            elif class_id == 2:
                ground_truth["shoes"] = ground_truth.get("shoes", 0) + 1
            elif class_id == 3:
                ground_truth["vest"] = ground_truth.get("vest", 0) + 1
            elif class_id == 4:
                ground_truth["worker"] = ground_truth.get("worker", 0) + 1
    return ground_truth

# Funzione per estrarre il campo "detections" dal JSON di previsione
def extract_json_from_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data.get("detections", {})


# Funzione per calcolare le metriche (TP, FP, FN)
def calculate_metrics(ground_truth, predictions, mapping):
    metrics = {"TP": 0, "FP": 0, "FN": 0}
    for class_name, model_key in mapping:
        pred_count = predictions.get(model_key, 0)
        true_count = ground_truth.get(class_name, 0)
        
        if pred_count == true_count:
            metrics["TP"] += true_count
        elif true_count > pred_count:
            metrics["TP"] += pred_count
            metrics["FN"] += (true_count - pred_count)
        else:
            metrics["TP"] += true_count
            metrics["FP"] += (pred_count - true_count)
    return metrics

def calculate_metrics_per_DPI(ground_truth, predictions, mapping, dpi):
    metrics = {"TP": 0, "FP": 0, "FN": 0}
    for class_name, model_key in mapping:
        if model_key is None or dpi != model_key.strip(":"):
            continue  # Skip if the current class does not match the DPI we're focusing on
        
        pred_count = predictions.get(model_key, 0)
        true_count = ground_truth.get(class_name, 0)
        
        if pred_count == true_count:
            metrics["TP"] += true_count
        elif true_count > pred_count:
            metrics["TP"] += pred_count
            metrics["FN"] += (true_count - pred_count)
        else:
            metrics["TP"] += true_count
            metrics["FP"] += (pred_count - true_count)
    
    return metrics

# Funzione per calcolare accuracy, precision, recall, e F1
def calculate_classification_metrics(metrics):
    TP = metrics["TP"]
    FP = metrics["FP"]
    FN = metrics["FN"]
    
    accuracy = round(TP / (TP + FP + FN), 2) if (TP + FP + FN) > 0 else 0
    precision = round(TP / (TP + FP), 2) if (TP + FP) > 0 else 0
    recall = round(TP / (TP + FN), 2) if (TP + FN) > 0 else 0
    f1_score = round(2 * (precision * recall) / (precision + recall), 2) if (precision + recall) > 0 else 0
     
    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1_score": f1_score
    }

def calculate_final_metrics(all_metrics):
    total_accuracy = 0
    total_precision = 0
    total_recall = 0
    total_f1_score = 0
    n = len(all_metrics)
    # Somma i valori delle metriche per ciascun file
    for metric in all_metrics:
        total_accuracy += metric['accuracy']
        total_precision += metric['precision']
        total_recall += metric['recall']
        total_f1_score += metric['f1_score']

        
    # Calcola la media per ciascuna metrica
    final_metrics = {
        "final_accuracy": total_accuracy / n, 
        "final_precision": total_precision / n, 
        "final_recall": total_recall / n, 
        "final_f1_score": total_f1_score / n
    }
    return final_metrics


# Funzione per salvare i risultati delle metriche in un file JSON
def save_metrics_to_file(metrics, output_folder, file_name):
    output_path = os.path.join(output_folder, f"{file_name}_metrics.json")
    with open(output_path, 'w') as f:
        json.dump(metrics, f, indent=4)
```

```python
ground_truth_folder = "PPE_Detection 2.v1i.yolov8-obb_FINALE/test/labels"
predictions_folder = "output/YOLOv8/test_set_predictions_results"
output_folder = "output/YOLOv8//metrics_output"
output_folder_final = "output/YOLOv8/metrics_output_final_aggregated"
os.makedirs(output_folder, exist_ok=True)
os.makedirs(output_folder_final, exist_ok=True)
DPI = ['worker', 'helmet', 'gloves', 'vest', 'shoes']
gt_files = sorted(os.listdir(ground_truth_folder))
pred_files = sorted(os.listdir(predictions_folder)) 
all_metrics = []
for gt_file, pred_file in zip(gt_files, pred_files):
    gt_results = read_ground_truth(os.path.join(ground_truth_folder,gt_file))
    data = extract_json_from_file(os.path.join(predictions_folder, pred_file))
    #print(data)
    prediction_results = filter_greater_than_zero(data)
    metrics = calculate_metrics(gt_results, prediction_results, mapping)
    base_filename = os.path.splitext(pred_file)[0]
    final_metrics = calculate_classification_metrics(metrics)
    all_metrics.append(final_metrics)
    save_metrics_to_file(final_metrics, output_folder, base_filename)
final = calculate_final_metrics(all_metrics)
print(f' final metrics: {final}')
save_metrics_to_file(final, output_folder_final, "metrics")

#########################################################################################################
final_ppes_metrics = {
    'worker': {
        'Accuracy': 0, 
        'Precision': 0, 
        'Recall': 0, 
        'F1-Score': 0
    },
    'helmet': {
        'Accuracy': 0, 
        'Precision': 0, 
        'Recall': 0, 
        'F1-Score': 0
    },
    'gloves': {
        'Accuracy': 0, 
        'Precision': 0, 
        'Recall': 0, 
        'F1-Score': 0
    },
    'vest': {
        'Accuracy': 0, 
        'Precision': 0, 
        'Recall': 0, 
        'F1-Score': 0
    },
    'shoes': {
        'Accuracy': 0, 
        'Precision': 0, 
        'Recall': 0, 
        'F1-Score': 0
    },
}
# Second part of the code to calculate per DPI metrics
gt_files = sorted(os.listdir(ground_truth_folder))
pred_files = sorted(os.listdir(predictions_folder)) 

# Iterate over each DPI type
for dpi in DPI:
    all_metrics_per_dpi = []  # Initialize a list to collect metrics for each DPI
    for gt_file, pred_file in zip(gt_files, pred_files):
        base_filename = os.path.splitext(pred_file)[0]

        # Read ground truth and predictions for the current image
        gt_results = read_ground_truth(os.path.join(ground_truth_folder, gt_file))
        data = extract_json_from_file(os.path.join(predictions_folder, pred_file))
        
        # Filter predictions with counts greater than zero
        prediction_results = filter_greater_than_zero(data)
        
        # Calculate per-image metrics for the current DPI
        metrics = calculate_metrics_per_DPI(gt_results, prediction_results, mapping, dpi)
        final_metrics_per_image = calculate_classification_metrics(metrics)
        all_metrics_per_dpi.append(final_metrics_per_image)

    # Calculate final averaged metrics for the current DPI
    final_dpi_metrics = calculate_final_metrics(all_metrics_per_dpi)
    print(f'final metric {dpi}: {final_dpi_metrics}')
    
    # Update the final PPE metrics dictionary with calculated values
    final_ppes_metrics[dpi]['Accuracy'] = final_dpi_metrics['final_accuracy']
    final_ppes_metrics[dpi]['Precision'] = final_dpi_metrics['final_precision']
    final_ppes_metrics[dpi]['Recall'] = final_dpi_metrics['final_recall']
    final_ppes_metrics[dpi]['F1-Score'] = final_dpi_metrics['final_f1_score']

# Save final PPEs metrics to file
output_file_path = os.path.join(output_folder_final, "final_ppes_metrics.json")
with open(output_file_path, 'w') as f:
    json.dump(final_ppes_metrics, f, indent=4)
```

```python
import json
import os

# Funzione per estrarre il tempo di elaborazione da un file JSON
def extract_processing_time(file_content):
    return file_content.get('processing_time_seconds', 0.0)

# Funzione per leggere il contenuto di un file JSON
def extract_json_from_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Directory delle predizioni
predictions_folder = "output/YOLOv8/test_set_predictions_results"
pred_files = sorted(os.listdir(predictions_folder))

# Variabili per sommare i tempi e contare i file
total_processing_time = 0
file_count = 0

# Itera su tutti i file di predizione per calcolare il tempo medio di elaborazione
for pred_file in pred_files:
    # Estrai i dati dal file JSON
    file_content = extract_json_from_file(os.path.join(predictions_folder, pred_file))
    
    # Estrai il tempo di elaborazione dal file
    processing_time = extract_processing_time(file_content)
    
    # Aggiungi il tempo di elaborazione al totale
    total_processing_time += processing_time
    file_count += 1

# Calcola la media del tempo di elaborazione
average_processing_time = total_processing_time / file_count if file_count > 0 else 0

# Stampa il risultato
print(f'Tempo medio di elaborazione per immagine: {average_processing_time:.6f} secondi')
```
